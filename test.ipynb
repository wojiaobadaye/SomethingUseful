```python
from sympy import symbols
import sympy
```

```python
x,t= symbols('x t')
A,B,C,f= symbols('A B C f', function=True)
```

```python
f = 62.5/(1+x/2)**t
A = sympy.summation(f, (t, 1, 11))+1000/(1+x/2)**11
A
```

```python
f = 50.625/(1+x/2)**t
C = sympy.summation(f, (t, 1, 16))+1000/(1+x/2)**16
C
```

```python
sympy.plot((A,(x, 0.01, 0.3)), (B,(x, 0.01, 0.3)), (C,(x, 0.01, 0.3)), xlim=(0.01, 0.3), ylim=(300, 2000), axis_center=(0.01, 300),label={'A':'a','B':'B','C':'C'},legend=True)
```
