```python
# 创建EdgeOptions对象并配置自动下载
options = webdriver.EdgeOptions() 
options.add_experimental_option("prefs", {
    "download.default_directory": dir_path,  # 指定下载文件保存的文件夹路径
    "download.prompt_for_download": False,  # 禁止弹窗询问下载
    "plugins.always_open_pdf_externally": True  # 总是在浏览器外部打开PDF文件
})
```

```python
# 定义浏览器
def set_broswer(driver_type = 'edge', path = broswer_path, options=options):
    if driver_type == 'edge':
        driver = webdriver.Edge(service=Service(path), options=options)
    elif driver_type == 'chrome':
        driver = webdriver.Chrome(path, options=options)
    else:
        raise ValueError (" use edge or chrome!")
    return driver
```

```python
# pdf合并方法
a = '15828440-41SA24H120N008R22G发票A901.pdf'
b = '15828440-41SA24H120N008R22G合同签收单物流单A901M.pdf'
# 指定待合并的PDF文件所在的目录
pdf_directory = '.'

# 获取目录下所有的PDF文件
pdf_files = [file for file in os.listdir(pdf_directory) if file.endswith('.pdf')]

# 创建一个PdfFileMerger对象
merger = PdfMerger()

# 遍历所有的PDF文件
for file_name in pdf_files:
    file_path = os.path.join(pdf_directory, file_name)
    
    # 将每个PDF文件添加到PdfFileMerger对象中
    merger.append(file_path)

# 指定合并后的PDF文件名
output_file = 'merged.pdf'

# 将合并后的PDF文件保存到指定路径
merger.write(output_file)

# 关闭PdfFileMerger对象
merger.close()

print(f'PDF合并完成！合并后的文件保存为 {output_file}')
```
