```python
def missing(df):
    missing_number = df.isnull().sum().sort_values(ascending=False)
    missing_percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([missing_number, missing_percent], axis=1, keys=['missing_number', 'missing_percent'])
    return missing_data
```

```python
def resample(data, num):
    ind = list(data[data['语音通话整体满意度']==num].index)
    n = len(ind)-300
    sam = random.sample(ind,k=n)
    data = data.drop(sam,axis=0)
    return data

for i in [8,9,10]:
    d = resample(d,i)
```
